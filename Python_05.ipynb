Quiz Problem: Guess the Number – Keep Guessing Until You Get It Right
Problem Description

Write a Python program that stores the value 10 in a variable called computer.
Your program should repeatedly ask the user to enter a number.

If the user’s guess is equal to computer, print "You gussed correct" and stop the loop.

If the guess is less than computer, print "Your number is too low".

If the guess is greater than computer, print "Your number is too high"

```python
computer=10

while True:
  n=int(input('Guess the number: '))
  if n < computer:
    print('Your number is too low')
  elif n > computer:
    print('Your number is too high')
  else:
    print('ou gussed correct')
    break
```

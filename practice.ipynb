```python
n=123
rev=0
while n > 0:
  r= (n%10)
  rev=rev*10+r
  n= (n//10)
rev2=0
while rev > 0:
  r=(rev%10)
  if r == 1:
    print('one')
  elif r ==2:
    print('two')
  elif r ==3:
    print('three')
  elif r ==4:
    print('four')
  elif r==5:
    print('five')
  rev2=(rev2*10+r)
  rev//=10
print(rev2)
```

```python
n=int(input('Enter the num '))
rev=0
sum=0
while n > 0:
  r= (n%10)
  rev=rev*10+r
  sum+=r
  n= (n//10)
print(rev)
print('Sum=',sum)
```

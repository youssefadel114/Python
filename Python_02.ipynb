****Task 2:** WAP to enter the age of 3 person and print oldest and yougest among them.**

```python
age1=int(input('Enter the 1st age: '))
age2=int(input('Enter the 2nd age: '))
age3=int(input('Enter the 3rd age: '))

the_oldest=max(age1, age2, age3)
the_youngest=min(age1, age2, age3)

output1='The first person is the oldest' if the_oldest == age1 else 'The second person is the oldest' if the_oldest == age2 else 'The third person is the oldest'
output2='The first person is the youngest' if the_youngest == age1 else 'The second person is the youngest' if the_youngest == age2 else 'The third person is the youngest'

print(output1)
print(output2)
```

**Task5: WAP simple calc that take two number from user and operator then apply oreration(+ - * / //).**

```python
numb1=int(input('The 1st number: '))
numb2=int(input('The 2nd number: '))
operator=input('Which operator u want (+ - * / //): ')

if operator == "+":
    print("Result=", numb1 + numb2)
elif operator == "-":
    print("Result=", numb1 - numb2)
elif operator == "*":
    print("Result=", numb1 * numb2)
elif operator == "/":
    print("Result=", numb1 / numb2)
elif operator == "//":
    print("Result=", numb1 // numb2)
else:
  print('Invalid')
```

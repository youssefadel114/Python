# Check If All Numbers Are Prime

Write a Python program to check if  number is prime . Return True if  number is prime otherwise False.


```python

l= [1,2,3,4,5,6,7]

for i in l:
  if i == 0 or i ==1:
    print(f'{i} -> False')
  elif i == 2:
    print(f'{i} -> True')
  else:
    for x in range(2, i):
      if i % x == 0:
        print(f'{i} -> false')
        break
      else:
        print(f'{i} -> True')
        break
```

# Difference Between Two Lists

Write a Python program to compute the difference between two lists.
Sample data: ["red", "orange", "green", "blue", "white"], ["black", "yellow", "green", "blue"]

Expected Output:

Color1-Color2: ['white', 'orange', 'red']

Color2-Color1: ['black', 'yellow']

```python
# i tried to do it with .pop, but it didn't work.

color1= ["red", "orange", "green", "blue", "white"]
color2= ["black", "yellow", "green", "blue"]


diff=[]
for i in color1:
  if i not in color2:
    diff.append(i)


diff2=[]
for i in color2:
  if i not in color1:
    diff2.append(i)

print('Color1-Color2=',diff)
print('Color2-Color1=',diff2)
```

# Reverse -ve number

Write a python code that reverse both negative and postive number

one code without built-in function

and

another within built-in function

```python
# Without build-in
l=[]
n=int(input('Enter the length of the list: '))
i=0
while i < n:
  num=int(input('Enter the number you want to add in your list: '))
  l.append(num)
  i+=1
l.sort()
l.reverse()
l
```

```python
# With build-in functions.
l=[]
n=int(input('Enter the length of the list: '))
i=0
while i < n:
  num=int(input('Enter the number you want to add in your list: '))
  l.append(num)
  i+=1
print(l)
l.sort()
x=list(reversed(l))
print(x)
```

Write a Python program that takes a product price and a discount percentage as input and prints the discount amount and the final price after discount.

**Example:**

Input:

* Price = 250
* Discount = 20



Output:

* Discount Amount = 50
* Final Price = 200

````python
x=float(input('Enter the price '))
y=float(input('Discount percentage? '))
discount_amount=(y/100)*x
final_price=(x-discount_amount)
print('Discount Amount=',discount_amount)
print('Final Price=',final_price)
````

Write a Python that takes two charachter as input from the user, then adds their ASCII values, and displays the result.

Search for bult-in function that get ascii code for any char.

````python
x= input("Enter the 1st character: ")
y= input("Enter the 2nd character: ")

a1= ord(x)
a2= ord(y)
summation= a1+a2

print(f"ASCII of '{x}' is: {a1}")
print(f"ASCII of '{y}' is: {a2}")
print(f"Sum of ASCII values = {summation}")
````

and

Write a Python program using input() that calculates the final percentage score, using the formula:

Then display the result on the screen.

````python
# Marks
m1=float(input('Enter the grade of the first exam '))
m2=float(input('Enter the grade of the second exam '))
m3=float(input('Enter the grade of the third exam '))

# Weights
w1=float(input('Enter the weighted average of the first exam : '))
w2=float(input('Enter the weighted average of the second exam : '))
w3=float(input('Enter the weighted average of the third exam : '))


final_score=((m1*w1+m2*w2+m3*w3)/100)

print('Your final score=',final_score)
````

```python
#	This function is called random module, it gets a random float number between 0 and 1.
import random
n = random.random()
print(n)
```

```python
# Is a build in function that can return a random value from a list, tuple and string.
import random
l = [5,3,2,5,7,9,0]
print(random.choice(l))
```

```python
# Generates a random number from a range.
print(random.randrange(20, 50))
```

```python
# An in-place function that can be used to change the position of a sequence.
import random

l= ['A', 'B', 'C', 'D', 'E']

random.shuffle(l)
print(l)
```

```python
# A function that gets a random value between 2 numbers. The lower limit (inclusive) and the upper limit (exclusive).
print(random.uniform(5, 10))
```

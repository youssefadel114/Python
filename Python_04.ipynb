Challenge: Write a program (WAP) that accepts a non-negative integer as input from the user and calculates its factorial using the concept described

Requirements:

Use a loop (like a for or while loop) to perform the repeated multiplication.

Handle the base case where the input is 0 (which should result in 1).

Display the final calculated factorial to the user.

The factorial is a mathematical function that finds the product of an integer and all the positive integers below it. It is denoted by the exclamation mark ($n!$), where $n$ is a non-negative integer.The fundamental idea of the factorial is the product of all positive integers from 1 up to $n$.Description:For any positive integer $n$, the factorial $n!$ is calculated by multiplying:$$\mathbf{n! = n \times (n-1) \times (n-2) \times \dots \times 3 \times 2 \times 1}$$Example: The factorial of 5 ($5!$) is:$$\mathbf{5! = 5 \times 4 \times 3 \times 2 \times 1 = 120}$$Special Case: The factorial of zero ($0!$) is defined as 1.

````python
x=int(input('Enter the number: '))
r=1
if x == 0:
  print('Factorial=',1)
else:
  while x > 0:
    r=r*x
    x-=1
  print('Factorial=',r)





````

An Armstrong number (also known as a narcissistic number, a pluperfect digital invariant (PPDI), or an amazing number)

is a number that is equal to the sum of its own digits raised to the power of the number of digits.

The Concept:Count the Digits (n): Determine how many digits the number has. This count is the power exponent.Calculate the Power Sum: Take each individual digit of the number and raise it to the power of $n$.

Check for Equality: Sum these results. If the final sum is equal to the original number, it is an Armstrong number.

Example: Let's check the number 153.Number of Digits (n): 153 has 3 digits, so $n=3$.Calculate the Power Sum:$$\mathbf{1^3 + 5^3 + 3^3}$$$$1 + 125 + 27 = 153$$Check: Since $153$ (original number) equals $153$ (sum of powers), 153 is an Armstrong number.

````python
n=int(input('Enter number '))
x=n
sum=0
while n > 0:
  r=n%10
  sum+=r**3
  n//=10

if sum==x :
  print(x,'is an Armstrong Number')
else:
  print(x,'is not an Armstrong Number')



````

````python
n=int(input('Enter number '))
rev=0
while n > 0:
  r=n%10
  rev=rev*10+r
  n//=10
rev2=0
sum=0

while rev >0:
  r=rev%10
  power=(r**3)
  sum+=power
  rev2=(rev2*10+r)
  rev//=10
if sum==rev2 :
  print(rev2,'is an Armstrong Number')
else:
  print(rev2,'is not an Armstrong Number')
````

### Write a program that asks the user to enter their name and age. The program should check if the user's age is greater than or equal to 18. ### Divisible by 7 and Multiples of 5

Write a Python program to find those numbers which are divisible by 7 and multiples of 5, between 1500 and 2700 (both included).

```python
i=int(input('Enter the min: '))
max=int(input('Enter the min: '))

if i >= max:
  print('Invalid')
else:
  while i <= max :
    if (i % 7 == 0) and (i % 5 ==0):
      print(i)
    i+=1
```

## Write a Python program that takes two integers from the user (start and end of a range), then counts how many even and odd numbers are in that range. Finally, print the total count of even and odd numbers.

```python
count_even=0
count_odd=0
i=int(input('Enter the min'))
max=int(input('Enter the min'))

while i <= max :
  if i % 2 == 0:
    print('Even')
    count_even+=1
  else:
    print('Odd')
    count_odd+=1
  print(i)
  i+=1
print('We have',count_even,'even number/s')
print('We have',count_odd,'odd number/s')
```

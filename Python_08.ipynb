# Square all element in the same list

# Find Second Largest Number in List

Write a Python program to find the second largest number in a list usig string methods or built in .

```python
l=[1,2,3,4,10,5,11,6,11]
l.remove(max(l))
max(l)
```

```python
l= [1,2,3,4,10,5,11,6,11]
unique= []

for i in l:
  if i not in unique:
    unique.append(i)
unique.sort()
print(f'The second largest num is: {unique[-2]}')
```
